Passo 3: Calcular as distancias entre cada GIS

```python
#importar Distancias o great_circle desde geopy
from geopy import distance
#pip install geopy

```

```python
#calcular as distancias para  2 direções
distance.distance(df.iloc[0][5],df.iloc[2][5])
```

```python
#calcular as distancias para todas as direções
import numpy as np

distancia=[]

for n in range(len(df)):
    for i in range(len(df)):
        if n!=i:
            distancia.append(np.round(distance.distance(df.iloc[n][5],df.iloc[i][5]).km,decimals=1))
            
print(distancia)
```

```python
 #criar "cidades" de acordo com o número de entradas do DataFrame
    
cidades=[i for i in range(len(df))]
cidades
```

```python
#criar os arcos
arcos=[(i,j) for i in cidades for j in cidades if i!=j]
print(arcos)
```

```python
#assinar as distancias a cada arco, utilizando um dicionario
distancia_arcos={(i,j):distance.distance(df.iloc[i][5],df.iloc[j][5]).km for i,j in arcos}
distancia_arcos
```

Passo 4: Resolver o Problema

```python
#importar Cplex
from docplex.mp.model import Model
```

```python
#criar o Modelo en Cplex
mdl=Model('TSP-2')
```

```python
#criar as variaveis de decisão em Cplex
x=mdl.binary_var_dict(arcos,name='x')
d=mdl.continuous_var_dict(cidades,name='d')
```

```python
#função objetivo do TSP

mdl.minimize(mdl.sum(distancia_arcos[i]*x[i] for i in arcos))
```

```python
#restrição do TSP
for c in cidades:
    mdl.add_constraint(mdl.sum(x[(i,j)] for i,j in arcos if i==c)==1, ctname='out_%d'%c)
```

```python
for c in cidades:
    mdl.add_constraint(mdl.sum(x[(i,j)] for i,j in arcos if j==c)==1, ctname='in_%d'%c)
```

```python
for i,j in arcos:
    if j!=0:
        mdl.add_indicator(x[(i,j)],d[i]+1==d[j],name='order_(%d,_%d)'%(i,j))
```

```python
print(mdl.export_to_string())
```

```python
#resolver o modelo
mdl.parameters.timelimit=60
mdl.parameters.mip.strategy.branch=1

solucion=mdl.solve(log_output=True)
```

```python
#revisar se a solução é ótima
mdl.get_solve_status()
```

```python
#obter a solução ótima e o valor da função objetivo
solucion.display()
```

```python
#[ [lat 1,log 1] , [lat 2,log 2] ]

coord_mapa=[]
for n in range(len(df)):
    lista_aux=[]
    lista_aux.append(df.iloc[n][5][0])
    lista_aux.append(df.iloc[n][5][1])
    
    coord_mapa.append(lista_aux)
    
print(coord_mapa)
```

```python
#importar a biblioteca

import folium
from folium import features

#pip install folium
```

```python
#graficar a solução

mapa=folium.Map(location=[coord_mapa[0][0],coord_mapa[0][1]])

for i,j in arcos_activos:
    linea=folium.PolyLine(locations=[[coord_mapa[i][0],coord_mapa[i][1]],[coord_mapa[j][0],coord_mapa[j][1]]],weight=5)
    
    mapa.add_child(linea)
    
    fg=folium.FeatureGroup(name="Direções")
    
    for n in range(len(df)):
        fg.add_child(folium.Marker(location=coord_mapa[n],popup=folium.Popup(nomes[n]),
                                    icon=folium.Icon(color='blue',icon_color='white',icon='info-sing')))
        
    mapa.add_child(fg)
    
mapa

```
